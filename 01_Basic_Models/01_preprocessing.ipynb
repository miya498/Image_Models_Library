# Image Preprocessing
このノートブックでは、基本的な画像前処理としてリサイズ、回転、反転、クロッピング、ノーマライズについて説明します。使用するサンプル画像は`Lenna.jpg`です。

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# サンプル画像の読み込み
image_path = './sample_image/Lenna.jpg'
image = cv2.imread(image_path, cv2.IMREAD_COLOR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.title("Original Image")
plt.axis('off')
plt.show()
```

## 1. リサイズ (Resizing)

```python
# リサイズ
resized_image = cv2.resize(image, (100, 100))

# 結果表示
plt.imshow(resized_image)
plt.title("Resized Image (100x100)")
plt.axis('off')
plt.show()
```

## 2. 回転 (Rotation)

```python
# 回転
height, width = image.shape[:2]
rotation_matrix = cv2.getRotationMatrix2D((width / 2, height / 2), 45, 1)
rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))

# 結果表示
plt.imshow(rotated_image)
plt.title("Rotated Image (45 degrees)")
plt.axis('off')
plt.show()
```

## 3. 反転 (Flipping)

```python
# 上下反転
flipped_image = cv2.flip(image, 0)

# 結果表示
plt.imshow(flipped_image)
plt.title("Flipped Image (Vertical)")
plt.axis('off')
plt.show()
```

## 4. クロッピング (Cropping)

```python
# クロッピング
cropped_image = image[50:200, 50:200]

# 結果表示
plt.imshow(cropped_image)
plt.title("Cropped Image")
plt.axis('off')
plt.show()
```

## 5. ノーマライズ (Normalization)

```python
# ノーマライズ
normalized_image = cv2.normalize(image, None, 0, 255, cv2.NORM_MINMAX)

# 結果表示
plt.imshow(normalized_image)
plt.title("Normalized Image")
plt.axis('off')
plt.show()
```
