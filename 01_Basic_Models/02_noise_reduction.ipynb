# Noise Reduction
このノートブックでは、画像のノイズ除去としてGaussian Blur、Median Blur、Bilateral Filterについて説明します。使用するサンプル画像は`Lenna.jpg`です。

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# サンプル画像の読み込み
image_path = './sample_image/Lenna.jpg'
image = cv2.imread(image_path, cv2.IMREAD_COLOR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.title("Original Image")
plt.axis('off')
plt.show()
```

## 1. Gaussian Blur
ガウシアンブラーは、ガウス分布に基づいたカーネルを使用して画像を平滑化し、ノイズを低減する手法です。周囲のピクセル値に基づき、中心ピクセルに重み付けを行い、エッジを比較的維持しながらぼかしを加えます。

```python
# Gaussian Blurを適用
gaussian_blurred = cv2.GaussianBlur(image, (5, 5), 0)

# 結果表示
plt.imshow(gaussian_blurred)
plt.title("Gaussian Blurred Image")
plt.axis('off')
plt.show()
```

## 2. Median Blur
メディアンブラーは、指定された領域のピクセルの中央値を使用してノイズを除去する方法です。特に「塩・コショウノイズ」と呼ばれる白黒の斑点状のノイズに効果的です。エッジの保持にも優れており、画像の細部を損ないにくい特徴があります。

```python
# Median Blurを適用
median_blurred = cv2.medianBlur(image, 5)

# 結果表示
plt.imshow(median_blurred)
plt.title("Median Blurred Image")
plt.axis('off')
plt.show()
```

## 3. Bilateral Filter
バイラテラルフィルタは、空間距離と色の類似度に基づいて重み付けを行い、エッジを保持しながらノイズを低減する高度な手法です。空間的な距離と色の類似度の2つの要素を考慮し、異なる色のエッジをぼかさずに、同じ色の領域のみをぼかします。

```python
# Bilateral Filterを適用
bilateral_filtered = cv2.bilateralFilter(image, 9, 75, 75)

# 結果表示
plt.imshow(bilateral_filtered)
plt.title("Bilateral Filtered Image")
plt.axis('off')
plt.show()
```
