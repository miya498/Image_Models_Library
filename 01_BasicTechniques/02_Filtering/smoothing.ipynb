# Smoothing
このノートブックでは、平均フィルタ、ガウシアンブラー、モーションブラーを用いた平滑化を行います。

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 画像読み込み
image_path = './sample_image/Lenna.jpg'
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

## 1. 平均フィルタ (Average Filter)
平均フィルタは、指定されたカーネルサイズ内のピクセルの平均値を計算し、中央のピクセルにその平均値を適用することでぼかしを加える手法です。最も基本的な平滑化手法であり、全体的に均一なぼかしが得られます

```python
# 平均フィルタ
average_blur = cv2.blur(image, (5, 5))

# 結果表示
plt.figure(figsize=(5, 5))
plt.title("Average Blur")
plt.imshow(average_blur)
plt.axis('off')
plt.show()
```

## 2. ガウシアンブラー (Gaussian Blur)
ガウシアンブラーは、ガウス分布に基づいたカーネルを使用して、より自然なぼかしを実現する平滑化手法です。平均フィルタと異なり、中央のピクセルほど重みが大きくなり、周囲のピクセルがぼかしに与える影響が減少するため、滑らかなぼかし効果を得ることができます。

```python
# ガウシアンブラー
gaussian_blur = cv2.GaussianBlur(image, (5, 5), 0)

# 結果表示
plt.figure(figsize=(5, 5))
plt.title("Gaussian Blur")
plt.imshow(gaussian_blur)
plt.axis('off')
plt.show()
```

## 3. モーションブラー (Motion Blur)
モーションブラーは、動きの軌跡に沿ってぼかしを適用する手法で、移動中の被写体や動感を表現する際に使われます。カーネルの形状を一方向に並べた形にすることで、移動方向にぼかしがかかります。

```python
# モーションブラー
kernel = np.zeros((15, 15))
kernel[:, 7] = 1 / 15
motion_blur = cv2.filter2D(image, -1, kernel)

# 結果表示
plt.figure(figsize=(5, 5))
plt.title("Motion Blur")
plt.imshow(motion_blur)
plt.axis('off')
plt.show()
```
