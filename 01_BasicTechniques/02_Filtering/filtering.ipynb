# Filtering
このノートブックでは、シャープニング、高周波フィルタ、低周波フィルタを用いた画像処理を行います。

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 画像読み込み
image_path = './sample_image/Lenna.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

## 1. シャープニング (Sharpening)
シャープニングは、画像のエッジや細かい部分を強調し、より鮮明に見せるためのフィルタです。隣接するピクセルとの差を強調することで、輪郭がはっきりとした画像を得ることができます。

```python
# シャープニングフィルタ
kernel_sharpening = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
sharpened = cv2.filter2D(image, -1, kernel_sharpening)

# 結果表示
plt.figure(figsize=(5, 5))
plt.title("Sharpened Image")
plt.imshow(sharpened, cmap='gray')
plt.axis('off')
plt.show()
```

## 2. 高周波フィルタ (High-Pass Filter)
高周波フィルタは、画像のエッジや急激な輝度変化を強調するフィルタです。エッジ検出にも利用され、画像の中で急激に変化する部分（高周波成分）を強調します。

```python
# 高周波フィルタ
kernel_high_pass = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])
high_pass_filtered = cv2.filter2D(image, -1, kernel_high_pass)

# 結果表示
plt.figure(figsize=(5, 5))
plt.title("High-Pass Filtered Image")
plt.imshow(high_pass_filtered, cmap='gray')
plt.axis('off')
plt.show()
```

## 3. 低周波フィルタ (Low-Pass Filter)
低周波フィルタは、画像の低周波成分を強調し、ノイズを低減するために使用されるフィルタです。画像全体を均一にぼかし、細かいノイズや高周波成分を抑制します。

```python
# 低周波フィルタ
kernel_low_pass = np.ones((5, 5), np.float32) / 25
low_pass_filtered = cv2.filter2D(image, -1, kernel_low_pass)

# 結果表示
plt.figure(figsize=(5, 5))
plt.title("Low-Pass Filtered Image")
plt.imshow(low_pass_filtered, cmap='gray')
plt.axis('off')
plt.show()
```
